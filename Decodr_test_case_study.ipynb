```python
data= data.drop(['school', 'address', 'famsize', 'Pstatus', 'reason', 'guardian', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'freetime'],axis=1)
```

```python
data.info()
```

```python
data= data.drop(["G1","G2"],axis=1)
```

```python
category_data= data.select_dtypes(include=["object"])
category_data
```

```python
dummy_data= pd.get_dummies(category_data)
dummy_data
```

```python
data= data.drop(category_data,axis=1)
data= pd.concat([data,dummy_data],axis=1)
data
```

```python
data.info()
```

```python
data.shape
```

```python
X= data.drop("G3",axis=1)
X
```

```python
y= data["G3"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3, random_state= 101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(max_depth=5)
```

```python
model= classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
